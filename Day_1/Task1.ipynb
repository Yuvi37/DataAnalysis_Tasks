```python
# Step 1: Load dataset with tab delimiter

df = pd.read_csv(r"C:\Users\yuvar\OneDrive - iitkgp.ac.in\Desktop\Elevate_DataAnalysis\Day 1\marketing_campaign.csv", sep='\t')

df.head()
```

```python
# Step 2: Overview
df.shape
df.info()
df.head()
```

```python
# Step 3: Handle missing values
missing = df.isnull().sum()
print("Missing values:\n", missing)
```

```python
# If only a few rows are missing Income, drop them (else we can impute)
df = df.dropna(subset=['Income'])
```

```python
# Count duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
# Step 5: Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Step 6: Convert date column
df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True)

# Step 7: Standardize text columns
df['education'] = df['education'].str.strip().str.title()
df['marital_status'] = df['marital_status'].str.strip().str.title()
```

```python
# Step 9: Data type checks
df.dtypes
```

```python
# Step 9: Export cleaned data
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
